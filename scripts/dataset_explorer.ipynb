```python
import os
import pandas as pd
```

```python
ROOT_DIR = os.path.dirname(os.getcwd())

ROOT_DIR
```

```python
df_raw = pd.read_csv(os.path.join(ROOT_DIR, 'datasets', 'gym_exercise_dataset.csv'))

df_raw.head()
```

```python
muscle_groups = df_raw[['Target_Muscles', 'Secondary Muscles']]

muscle_groups.head()
```

```python
muscle_groups['full_set'] = muscle_groups.apply(
    lambda x: x['Target_Muscles'] + (x['Secondary Muscles'] if isinstance(x['Secondary Muscles'], str) else ''),
    axis=1)

muscle_groups.head()
```

```python
list_muscles = [group.split(',') for group in muscle_groups['full_set'].tolist()]

list_muscles = [muscle.strip() for group in list_muscles for muscle in group]

set(list_muscles)
```
