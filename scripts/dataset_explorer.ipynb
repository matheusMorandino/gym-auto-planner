```python
import os
import pandas as pd
```

```python
ROOT_DIR = os.path.dirname(os.getcwd())

ROOT_DIR
```

```python
df_raw = pd.read_csv(os.path.join(ROOT_DIR, 'datasets', 'gym_exercise_dataset.csv'))

df_raw.head()
```

```python
df_raw.columns
```

```python
muscle_groups = df_raw[['Target_Muscles', 'Synergist_Muscles', 'Stabilizer_Muscles', 'Antagonist_Muscles', 'Dynamic_Stabilizer_Muscles', 'Secondary Muscles']]

muscle_groups.head()
```

```python
muscle_groups['full_set'] = muscle_groups.fillna('').astype(str).agg(','.join, axis=1).replace('\n', '', regex=True).replace(r'\(.*?\)', '', regex=True).replace(r'[0-9]', '', regex=True).replace('None', '')


muscle_groups.head()
```

```python
list_muscles = [group.split(',') for group in muscle_groups['full_set'].tolist()]

list_muscles = [muscle.strip() for group in list_muscles for muscle in group]

set(list_muscles)
```

```python
muscle_groups['target_count'] = muscle_groups.apply(lambda x: len([m for m in x['Target_Muscles'].split(',') if len(m.strip()) > 0]), axis=1)

muscle_groups[muscle_groups['target_count'] > 1]
```
